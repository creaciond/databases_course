```python
import pandas as pd
import re
```

## Объединение с очищенными данными

Изначальный датасет:

```python
df = pd.read_csv("./data/metadata.tsv", sep="\t", encoding="utf-8")
df.head()
```

Данные после OpenRefine:

```python
df_refined = pd.read_csv("./data/Tolstoy-metadata-OR.tsv", sep="\t", encoding="utf-8")
df_refined.head()
```

Объединение:

```python
df["AUTHOR_REFINED"] = df_refined["AUTHOR_REFINED"].values
df["GEOGR_REFINED"] = df_refined["GEOGR_REFINED"].values
```

```python
df.head()
```

## Преобразование для загрузки в БД

Кодирование для отдельных таблиц:

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

### Место съёмки

```python
df["GEOGR_REFINED"] = le.fit_transform(df["GEOGR_REFINED"])
```

```python
with open("db_locations.tsv", "w", encoding="utf-8") as f:
    f.write("idLocation\tlocationName\n")
    for i, loc in enumerate(le.classes_):
        f.write(f"{i}\t{le.classes_[i]}\n")
```

### Фотограф

```python
df["AUTHOR_REFINED"] = le.fit_transform(df["AUTHOR_REFINED"])
```

```python
with open("db_authors.tsv", "w", encoding="utf-8") as f:
    f.write("idAuthor\tauthorName\n")
    for i, loc in enumerate(le.classes_):
        f.write(f"{i}\t{le.classes_[i]}\n")
```

### Рубрики

```python
i = 0
overall_spirubr = {}
for item in df.iterrows():
    for spirubr in item[1]["SPIRUBR"].split("; "):
        if spirubr not in overall_spirubr.keys():
            overall_spirubr[spirubr] = i
            i += 1
```

Соответствие рубрик и их ID:

```python
with open("db_rubrics.tsv", "w", encoding="utf-8") as f:
    f.write("idRubric\trubricName\n")
    for i, key in enumerate(overall_spirubr.keys()):
        f.write(f"{overall_spirubr[key]}\t{key}\n")
```

Рубрики на каждой фотографии:

```python
with open("db_rubrics_photos.tsv", "w", encoding="utf-8") as f:
    f.write("idPhoto\tidRubric\n")
    for item in df.iterrows():
        id_photo = item[1]["id"]
        for rubric in item[1]["SPIRUBR"].split("; "):
            f.write(f"{id_photo}\t{overall_spirubr[rubric]}\n")
```

### Основное описание фотографии

```python
year_regex = re.compile("1[0-9]{3}")
df["year"] = df["CREAT"].apply(lambda x: int(year_regex.search(x).group(0)) if year_regex.search(x) else -1)
```

```python
photo_desc = df[["id", "AUTHOR_REFINED", "GEOGR_REFINED", "year", "COMPNAM"]]
photo_desc.to_csv("db_photo_descriptions.tsv", sep="\t", encoding="utf-8", index=None)
```

### Музейные пометы

```python
museum_annotations = df[["id", "IDENTIF", "INV", "AUTHOR", "GEOGR"]]
museum_annotations.to_csv("db_museum_annotations.tsv", sep="\t", encoding="utf-8", index=None)
```

### Описание самой фотографии физическое

```python
dimensions = re.compile("([0-9,]+) ?х ?([0-9,]+)")
df["width"] = df["SIZES"].apply(lambda x: float(dimensions.search(x).group(1).replace(",", ".")) if dimensions.search(x) else -1)
df["height"] = df["SIZES"].apply(lambda x: float(dimensions.search(x).group(2).replace(",", ".")) if dimensions.search(x) else -1)
```

```python
physical_descr = df[["id", "width", "height", "INSCR"]]
physical_descr.to_csv("db_physical_descr.tsv", sep="\t", encoding="utf-8", index=None)
```

### Файлы

```python
files = df[["id", "miniature_path", "fullsize_path"]]
files.to_csv("db_files.tsv", sep="\t", encoding="utf-8", index=None)
```
